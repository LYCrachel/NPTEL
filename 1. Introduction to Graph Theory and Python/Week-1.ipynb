## Lecture 06 - Introduction to Networkx-2

```python
#Design the a networrk representing the roads between the cities in India.
import networkx as nx
import matplotlib.pyplot as plt
import random
from random import choice
```

```python
C =nx.Graph() #undirected graph
# G =nx.DiGraph() >>>> Directed graph
City_set =['Delhi','banglore','Hyderabad','Ahmedabad','Chennai','Kolkata','Surat','Pune','Jaipur']
for each in City_set:
    C.add_node(each)
nx.draw(C)
plt.show()
```

```python
costs =[]
value =100
while (value<=2000):
    costs.append(value)
    value =value+100
print costs
print len(costs)
print "Edges : ", C.number_of_edges()
print "Nodes : ", C.number_of_nodes()
```

```python
#We are going to add 1 edges to this network

print(C.nodes())
#Addding 10 edges to the graph
while (C.number_of_edges()<16):
    city1 = random.choice(City_set)
    city2 = random.choice(City_set)
    wt = random.choice(costs)#selecting random costs
    if city1 != city2 and C.has_edge(city1, city2) == 0:
        C.add_edge(city1, city2, weight = wt)
nx.draw(C, with_labels=1)
plt.show()
```
